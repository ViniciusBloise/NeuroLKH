```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import os
import re
```

```python
model_cmp = 'result/instances/stats/cmp-vsr-neuro-pi.csv'
header = ['model', 'dimension', 'cand.set.type', 'trials/timelimit', 'trials max','trials eff', 'bestrun', 'time']
df = pd.read_csv(model_cmp, names=header)
```

```python
df.head()
```

```python
df.columns
```

```python
dim = 5000
trials = df.query(f"`trials/timelimit`=='trials' and dimension=={dim}")

g_res = sns.lineplot(data=trials, x='trials max', y='bestrun', hue='model' )
g_res.set(xscale='log')
g_res.set_title(f'Dimension = {dim}')
```

```python
g_res = sns.lineplot(data=trials, x='Trials max', y='BestRun', hue='VSR/Neuro' )
g_res.set(xscale='log')
```

```python
timelimit = df.query("`Trials/TimeLimit`=='timelimit'")

g_res = sns.lineplot(data=timelimit, x='Trials max', y='BestRun', hue='VSR/Neuro' )
g_res.set(xscale='log')
```
